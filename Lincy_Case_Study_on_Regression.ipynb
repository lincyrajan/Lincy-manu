```python
x=df.drop(["Price"],axis=1)
x
```

```python
y=df['Price']
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

Building a Linear Model

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
lm=LinearRegression()
lm_model=lm.fit(x_train,y_train)
lm_predict=lm_model.predict(x_test)
```

```python
lm_model=lm.fit(x_train,y_train)
lm_predict=lm_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print("Mean Squared Error =",mean_squared_error(y_test,lm_predict))
print("R Squared Error=",r2_score(y_test,lm_predict))
```

Lasso Regression model

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.1)
lm_model=lasso.fit(x_train,y_train)
lm_predict=lm_model.predict(x_test)
```

```python
print("Mean Squared Error=",mean_squared_error(y_test,lm_predict))
print("R Squared Error=",r2_score(y_test,lm_predict))
```

Prediction of the Price in the year 2022(Using Linear Regression)

```python
x_2022=pd.DataFrame({'Year':[2022]})
```

```python
lm_predict=lm_model.predict(x_2022)
```
